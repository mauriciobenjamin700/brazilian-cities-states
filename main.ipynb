```python
import chardet

def convert_to_utf8(input_string):
    # Detectar a codificação da string
    detected_encoding = chardet.detect(input_string)['encoding']
    
    # Converter a string para UTF-8
    utf8_string = input_string.decode(detected_encoding).encode('utf-8')
    
    return utf8_string.decode('utf-8')

#Exemplo de uso
original_string = b"ALTO GAR\xc7AS"  # String em Latin-1
utf8_string = convert_to_utf8(original_string)

# Converter a string UTF-8 de volta para uma string normal

print(utf8_string)  # Deve imprimir "ALTO GARÇAS"
```

```python
def df_to_dict(df: pd.DataFrame) -> dict:
    state = str(df["SG_UF"].unique()[0])
    cities = df["NM_MUNICIPIO"].unique().tolist()
    cities = [str(city) for city in cities]
    #cities = [convert_to_utf8(city) for city in cities]
    return {state: cities}

data_dict = [df_to_dict(df) for df in dfs]
data_dict
```

```python
from collections import defaultdict

def merge_dicts(dict_list):
    merged_dict = defaultdict(list)
    
    for d in dict_list:
        for key, value in d.items():
            merged_dict[key].extend(value)
    
    return dict(merged_dict)

# Exemplo de uso
# dict1 = {
#     "MT": ["ACORIZAL", "ALTA FLORESTA"],
#     "SP": ["SÃO PAULO", "CAMPINAS"]
# }

# dict2 = {
#     "MT": ["ALTO ARAGUAIA", "ALTO BOA VISTA"],
#     "RJ": ["RIO DE JANEIRO", "NITERÓI"]
# }

# dict3 = {
#     "SP": ["SOROCABA"],
#     "RJ": ["PETRÓPOLIS"]
# }

# dict_list = [dict1, dict2, dict3]
# merged_dict = merge_dicts(dict_list)

# print(merged_dict)
```
